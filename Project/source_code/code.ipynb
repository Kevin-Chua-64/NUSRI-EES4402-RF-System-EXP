### a.
***If 2-PAM and 4-PAM have the same average bit energy $E_b$, indicate the symbols in vector form for 2-PAM and 4-PAM using $E_b$.***

$\quad$The signal function is $s_m(t)=A_mp(t)$, where $A_m=2m-1-M, 1\le m\le M$. $p(t)$ is the pulse with duration $T_s$, and denote $\epsilon_p=\int_{-\infty}^{\infty}p^2(t)dt$.

$\quad$Since PAM has one dimension, the complete orthonormal basis set is $\varphi(t)=\frac{p(t)}{\sqrt{\epsilon_p}}$. The vector form is $S_m=A_m\sqrt{\epsilon_p}$.

$\quad$The average energy is $E_{avg}=\frac{1}{M}\sum_{m=1}^M A_m^2\epsilon_p=\frac{\epsilon_p}{M}\sum_{m=1}^M A_m^2=\frac{(M^2-1)\epsilon_p}{3}$. The average bit energy is $E_b=\frac{E_{avg}}{\log_{2}{M}}=\frac{(M^2-1)\epsilon_p}{3\log_{2}{M}}$.

$\quad$In the case that 2-PAM and 4-PAM have the same $E_b$, then $E_b=\frac{(2^2-1)\epsilon_{p2PAM}}{3\log_{2}{2}}=\frac{(4^2-1)\epsilon_{p4PAM}}{3\log_{2}{4}}$. Thus, $\epsilon_{p2PAM}=E_b$ and $\epsilon_{p4PAM}=\frac{2E_b}{5}$.

$\quad \bullet$ For 2-PAM, $S_1(0)=-\sqrt{E_b}, S_2(1)=\sqrt{E_b}$.

$\quad \bullet$ For 4-PAM, $S_1(00)=-3\sqrt{\frac{2E_b}{5}}, S_2(01)=-\sqrt{\frac{2E_b}{5}}, S_3(11)=\sqrt{\frac{2E_b}{5}}, S_4(10)=3\sqrt{\frac{2E_b}{5}}$.

### b.
***Derive the exact BER for 2-PAM and 4-PAM in terms of the signal-to-noise ratio (SNR) defined as $\frac{E_b}{N_0}$ when the Gray mapping is applied.***

$\quad$The decision metric:<br>
Denote the vector form of the received signal by $\vec{r}=R$. Since all symbols are equiprobable, apply the ML rule: $\hat{s}=\arg {\underset{1\le m\le M}{\min}}{\Vert \vec{r}-\vec{s_m}\Vert}$. More specific, PAM is one-dimensional, $\hat{s}=\arg {\underset{1\le m\le M}{\min}}|R-S_m|$.

$\quad$The exact BER:<br>
When the Gray mapping is applied, two adjacent symbols only have one bit changes. So, we can know the number of bit errors corresponding to each integration region, and specific the symbol error rate to bit error rate. Denote the bit SNR $\frac{E_b}{N_0}=\gamma_b$.

$\quad \bullet$ For 2-PAM, the decision metric:

$D_1=\{|R-S_1|\lt |R-S_2|\}=\{|R+\sqrt{E_b}|\lt |R-\sqrt{E_b}|\}=\{R\lt 0\}$<br>
$D_2=\{|R-S_2|\lt |R-S_1|\}=\{|R-\sqrt{E_b}|\lt |R+\sqrt{E_b}|\}=\{R\gt 0\}$

$\quad$The BER:

$P_b=\frac{1}{2}[\int_{0}^{\infty}P(r|s_1)dr+\int_{-\infty}^{0}P(r|s_2)dr]$<br>
$\quad$ $=\frac{1}{2}[\int_{0}^{\infty}P(n=r-s_1)dr+\int_{-\infty}^{0}P(n=r-s_2)dr]$<br>
$\quad$ $=\frac{1}{2}[\int_{0}^{\infty}\frac{1}{\sqrt{\pi N_0}}e^{-\frac{(r+\sqrt{E_b})^2}{N_0}}dr+\int_{-\infty}^{0}\frac{1}{\sqrt{\pi N_0}}e^{-\frac{(r-\sqrt{E_b})^2}{N_0}}dr]$<br>
$\quad$ $=\frac{1}{2}[\int_{\sqrt{\frac{2E_b}{N_0}}}^{\infty}\frac{1}{\sqrt{2\pi}}e^{-\frac{t^2}{2}}dt+\int_{-\infty}^{-\sqrt{\frac{2E_b}{N_0}}}\frac{1}{\sqrt{2\pi}}e^{-\frac{t^2}{2}}dt]$<br>
$\quad$ $=\frac{1}{2}\{Q(\sqrt{\frac{2E_b}{N_0}})+[1-Q(-\sqrt{\frac{2E_b}{N_0}})]\}$<br>
$\quad$ $=Q(\sqrt{\frac{2E_b}{N_0}})=Q(\sqrt{2\gamma_b})$

$\quad \bullet$ For 4-PAM, the decision metric:

$D_1=\{|R-S_1|\lt {\underset{m=2,3,4}{\min}}\{|R-S_m|\}\}=\{|R-S_1|\lt |R-S_2|\}=\{|R+3\sqrt{\frac{2E_b}{5}}|\lt |R+\sqrt{\frac{2E_b}{5}}|\}=\{R\lt -2\sqrt{\frac{2E_b}{5}}\}$<br>
$D_2=\{|R-S_2|\lt {\underset{m=1,3,4}{\min}}\{|R-S_m|\}\}=\{|R-S_2|\lt |R-S_1|\}\bigcup\{|R-S_2|\lt |R-S_3|\}=\{|R+\sqrt{\frac{2E_b}{5}}|\lt |R+3\sqrt{\frac{2E_b}{5}}|\}\bigcup\{|R+\sqrt{\frac{2E_b}{5}}|\lt |R-\sqrt{\frac{2E_b}{5}}|\}=\{-2\sqrt{\frac{2E_b}{5}}\lt R\lt 0\}$<br>
$D_3=\{|R-S_3|\lt {\underset{m=1,2,4}{\min}}\{|R-S_m|\}\}=\{|R-S_3|\lt |R-S_2|\}\bigcup\{|R-S_3|\lt |R-S_4|\}=\{|R-\sqrt{\frac{2E_b}{5}}|\lt |R+\sqrt{\frac{2E_b}{5}}|\}\bigcup\{|R-\sqrt{\frac{2E_b}{5}}|\lt |R-3\sqrt{\frac{2E_b}{5}}|\}=\{0\lt R\lt 2\sqrt{\frac{2E_b}{5}}\}$<br>
$D_4=\{|R-S_4|\lt {\underset{m=1,2,3}{\min}}\{|R-S_m|\}\}=\{|R-S_4|\lt |R-S_3|\}=\{|R-3\sqrt{\frac{2E_b}{5}}|\lt |R-\sqrt{\frac{2E_b}{5}}|\}=\{R\gt 2\sqrt{\frac{2E_b}{5}}\}$

$\quad$The BER: by symmetric, $P_{b|s_1}=P_{b|s_4}, P_{b|s_2}=P_{b|s_3}$.

When $S_1$ is sent, it has 1 bit error if $S_2$ or $S_4$ is received and 2 bits error if $S_3$ is received.

When $S_2$ is sent, it has 1 bit error if $S_1$ or $S_3$ is received and 2 bits error if $S_4$ is received.

$P_b=\frac{1}{2}[(\frac{1}{2}\int_{-2\sqrt{\frac{2E_b}{5}}}^{0}P(r|s_1)dr+\int_{0}^{2\sqrt{\frac{2E_b}{5}}}P(r|s_1)dr+\frac{1}{2}\int_{2\sqrt{\frac{2E_b}{5}}}^{\infty}P(r|s_1)dr)
+(\frac{1}{2}\int_{-\infty}^{-2\sqrt{\frac{2E_b}{5}}}P(r|s_2)dr+\frac{1}{2}\int_{0}^{2\sqrt{\frac{2E_b}{5}}}P(r|s_2)dr+\int_{2\sqrt{\frac{2E_b}{5}}}^{\infty}P(r|s_2)dr)]$<br>
$\quad$ $=\frac{1}{2}[(\frac{1}{2}\int_{-2\sqrt{\frac{2E_b}{5}}}^{0}\frac{1}{\sqrt{\pi N_0}}e^{-\frac{(r+3\sqrt{\frac{2E_b}{5}})^2}{N_0}}dr+\int_{0}^{2\sqrt{\frac{2E_b}{5}}}\frac{1}{\sqrt{\pi N_0}}e^{-\frac{(r+3\sqrt{\frac{2E_b}{5}})^2}{N_0}}dr+\frac{1}{2}\int_{2\sqrt{\frac{2E_b}{5}}}^{\infty}\frac{1}{\sqrt{\pi N_0}}e^{-\frac{(r+3\sqrt{\frac{2E_b}{5}})^2}{N_0}}dr)$<br>
$\quad$ $+(\frac{1}{2}\int_{-\infty}^{-2\sqrt{\frac{2E_b}{5}}}\frac{1}{\sqrt{\pi N_0}}e^{-\frac{(r+\sqrt{\frac{2E_b}{5}})^2}{N_0}}dr+\frac{1}{2}\int_{0}^{2\sqrt{\frac{2E_b}{5}}}\frac{1}{\sqrt{\pi N_0}}e^{-\frac{(r+\sqrt{\frac{2E_b}{5}})^2}{N_0}}dr+\int_{2\sqrt{\frac{2E_b}{5}}}^{\infty}\frac{1}{\sqrt{\pi N_0}}e^{-\frac{(r+\sqrt{\frac{2E_b}{5}})^2}{N_0}}dr)]$<br>
$\quad$ $=\frac{1}{2}[(\frac{1}{2}\int_{2\sqrt{\frac{E_b}{5N_0}}}^{6\sqrt{\frac{E_b}{5N_0}}}\frac{1}{\sqrt{2\pi}}e^{-\frac{t^2}{2}}dt+\int_{6\sqrt{\frac{E_b}{5N_0}}}^{10\sqrt{\frac{E_b}{5N_0}}}\frac{1}{\sqrt{2\pi}}e^{-\frac{t^2}{2}}dt+\frac{1}{2}\int_{10\sqrt{\frac{E_b}{5N_0}}}^{\infty}\frac{1}{\sqrt{2\pi}}e^{-\frac{t^2}{2}}dt)$<br>
$\quad$ $+(\frac{1}{2}\int_{-\infty}^{-2\sqrt{\frac{E_b}{5N_0}}}\frac{1}{\sqrt{2\pi}}e^{-\frac{t^2}{2}}dt+\frac{1}{2}\int_{2\sqrt{\frac{E_b}{5N_0}}}^{6\sqrt{\frac{E_b}{5N_0}}}\frac{1}{\sqrt{2\pi}}e^{-\frac{t^2}{2}}dt+\int_{6\sqrt{\frac{E_b}{5N_0}}}^{\infty}\frac{1}{\sqrt{2\pi}}e^{-\frac{t^2}{2}}dt)]$<br>
$\quad$ $=\frac{1}{2}\{\frac{1}{2}[Q(2\sqrt{\frac{E_b}{5N_0}})-Q(6\sqrt{\frac{E_b}{5N_0}})]+[Q(6\sqrt{\frac{E_b}{5N_0}})-Q(10\sqrt{\frac{E_b}{5N_0}})]+\frac{1}{2}Q(10\sqrt{\frac{E_b}{5N_0}})$<br>

### c.
***Simulate 2-PAM and 4-PAM modulation and detection in baseband with AWGN to find the experimental BER using Jupyter notebook or Matlab. Please simulate until the BER reaches $10^{âˆ’6}$ by varying the SNR. Plot the curve of BER vs. SNR. In this plot, the horizontal axis is SNR in dB-scale ($10\log_{10}{SNR}$), and the vertical axis is BER in log-scale ($10\log_{10}{BER}$).***

<img src="figs/flowgraph.png">

```python
import numpy as np
import random

# Perform 2-PAM modulation 0->(-1) 1->(1)
def pam2_gray_modulation(bits, Eb):
    symbols = -1 + 2*bits
    symbols = symbols * np.sqrt(Eb)
    return symbols

# Perform 4-PAM modulation 00->(-3) 01->(-1) 11->(1) 10->(3)
def pam4_gray_modulation(bits, Eb):
    symbols = (-1 + 2*bits[::2]) *  (3 - 2*bits[1::2])
    symbols = symbols * np.sqrt(0.4*Eb)
    return symbols

# Simulate AWGN channel N(0,N_0/2)
def pam_add_awgn_noise(signal, noise_power):
    noise = np.sqrt(noise_power/2) * np.random.randn(*signal.shape)
    received_signal = signal + noise
    return received_signal

# Perform 2-PAM demodulation D1={r<0}, D2={r>0}
def pam2_demodulation(received_symbols):
    demodulated_bits = np.zeros(len(received_symbols), dtype=int)
    demodulated_bits = received_symbols > 0
    demodulated_bits[received_symbols==0] = random.choices([0, 1], k=1)
    return demodulated_bits

# Perform 4-PAM demodulation D1={r<-2sqrt(0.4*Eb)}, D2={-2sqrt(0.4*Eb)<r<0}, D3={0<r<2sqrt(0.4*Eb)}, D4={r>2sqrt(0.4*Eb)}
def pam4_demodulation(received_symbols):
    demodulated_bits = np.zeros(2 * len(received_symbols), dtype=int)
    demodulated_bits[::2] = received_symbols > 0
    demodulated_bits[::2][received_symbols==0] = random.choices([0, 1], k=1)
    demodulated_bits[1::2] = (received_symbols > -2*np.sqrt(0.4*Eb)) & (received_symbols < 2*np.sqrt(0.4*Eb))
    demodulated_bits[1::2][(received_symbols == -2*np.sqrt(0.4*Eb)) | (received_symbols == 2*np.sqrt(0.4*Eb))] = random.choices([0, 1], k=1)
    return demodulated_bits

# Calculate error bit number
def error_num(original_bits, received_bits):
    return np.sum(original_bits != received_bits)


# Simulation parameters
Eb = 1.0  # Fixed bit energy
num_bits = 10000  # Number of bits per simulation
error_up_num = 100  # At least error bits to find
# SNR range in dB
pam2_snr_range_dB = np.append(np.arange(-30, 10.4, 0.2), np.arange(10.4, 11.05, 0.05))
pam4_snr_range_dB = np.append(np.arange(-30, 14.2, 0.2), np.arange(14.2, 15.05, 0.05))
# SNR range
pam2_snr_range = 10**(pam2_snr_range_dB/10)
pam4_snr_range = 10**(pam4_snr_range_dB/10)
# Noise range
pam2_noise_range = Eb / pam2_snr_range
pam4_noise_range = Eb / pam4_snr_range

# Initialize arrays to store BER values
ber_values_2pam = np.zeros(len(pam2_snr_range_dB))
ber_values_4pam = np.zeros(len(pam4_snr_range_dB))

# Perform simulation for each noise value (2-PAM)
def PAM2():
    for i, noise in enumerate(pam2_noise_range):
        error_count = 0
        loop_count = 0
        while(error_count <= error_up_num):  # Simulate until error bits is sufficient
            bits = np.random.randint(0, 2, num_bits)  # Generate random bits
            modulated_symbols = pam2_gray_modulation(bits, Eb)
            received_symbols = pam_add_awgn_noise(modulated_symbols, noise)
            demodulated_bits = pam2_demodulation(received_symbols)
            error_count += error_num(bits, demodulated_bits)
            loop_count += 1
            
        # Calculate BER
        ber_values_2pam[i] = 1.0*error_count / (num_bits*loop_count)


# Perform simulation for each noise value (4-PAM)
def PAM4():
    for i, noise in enumerate(pam4_noise_range):
        error_count = 0
        loop_count = 0
        while(error_count <= error_up_num):  # Simulate until error bits is sufficient
            bits = np.random.randint(0, 2, num_bits)  # Generate random bits
            modulated_symbols = pam4_gray_modulation(bits, Eb)
            received_symbols = pam_add_awgn_noise(modulated_symbols, noise)
            demodulated_bits = pam4_demodulation(received_symbols)
            error_count += error_num(bits, demodulated_bits)
            loop_count += 1
            
        # Calculate BER
        ber_values_4pam[i] = 1.0*error_count / (num_bits*loop_count)

# Simulation
PAM2()
PAM4()
```

### d.
***Plot the exact BER derived in step b in the same plot and comment on the result.***

```python
from scipy.special import erfc
import matplotlib.pyplot as plt

# Q function
def Qx(x):
    return 0.5 * erfc(np.sqrt(0.5) * x)

# The theoretical curve
ber_theorem_2pam = Qx(np.sqrt(2*pam2_snr_range))
ber_theorem_4pam = 0.75*Qx(np.sqrt(0.8*pam4_snr_range)) + 0.5*Qx(np.sqrt(7.2*pam4_snr_range)) - 0.25*Qx(np.sqrt(20*pam4_snr_range))


# Plot the curves
plt.figure(0, figsize=(10, 6))
plt.plot(pam2_snr_range_dB, np.log10(ber_values_2pam),  label='Simulation',  color='blue', marker='.')
plt.plot(pam2_snr_range_dB, np.log10(ber_theorem_2pam), label='Theoretical', color='red', linestyle='--')
pam2_reach_10_6_snr = pam2_snr_range_dB[np.argmin(np.abs(ber_values_2pam - 1e-6))]  # SNR when BER reaches 10^-6
plt.text(pam2_reach_10_6_snr-7, -6.25, r'BER:$10^{-6}$'+f'\nSNR:{pam2_reach_10_6_snr:.2f}dB', fontsize=10)
plt.xlim([-33, 17])
plt.ylim([-7, 0])
plt.title('2-PAM')
plt.xlabel('SNR(dB-scale)')
plt.ylabel('BER(log-scale)')
plt.legend()
plt.grid(True)

plt.figure(1, figsize=(10, 6))
plt.plot(pam4_snr_range_dB, np.log10(ber_values_4pam),  label='Simulation',  color='blue', marker='.')
plt.plot(pam4_snr_range_dB, np.log10(ber_theorem_4pam), label='Theoretical', color='red', linestyle='--')
pam4_reach_10_6_snr = pam4_snr_range_dB[np.argmin(np.abs(ber_values_4pam - 1e-6))]  # SNR when BER reaches 10^-6
plt.text(pam4_reach_10_6_snr-7, -6.25, r'BER:$10^{-6}$'+f'\nSNR:{pam4_reach_10_6_snr:.2f}dB', fontsize=10)
plt.xlim([-33, 17])
plt.ylim([-7, 0])
plt.title('4-PAM')
plt.xlabel('SNR(dB-scale)')
plt.ylabel('BER(log-scale)')
plt.legend()
plt.grid(True)

plt.figure(2, figsize=(10, 6))
plt.plot(pam2_snr_range_dB, np.log10(ber_theorem_2pam), label='2-PAM', color='blue', linestyle='--')
plt.plot(pam4_snr_range_dB, np.log10(ber_theorem_4pam), label='4-PAM', color='red', linestyle='--')
plt.plot([pam2_reach_10_6_snr,pam4_reach_10_6_snr], [-6,-6], color='green', linestyle='dashed', linewidth=3)
plt.text(pam2_reach_10_6_snr+0.5, -6.3, f'{pam4_reach_10_6_snr-pam2_reach_10_6_snr:.2f}dB', color='green', fontsize=10)
plt.xlim([-33, 17])
plt.ylim([-7, 0])
plt.title('2-PAM and 4-PAM Comparison (Theoretical)')
plt.xlabel('SNR(dB-scale)')
plt.ylabel('BER(log-scale)')
plt.legend()
plt.grid(True)

plt.show()
```

**Comment:**

$\quad$It can be seen that the simulation results of both 2-PAM and 4-PAM are very close to the theoretical values. The BER decreases as the SNR rises.

$\quad$For 2-PAM, the simulation result of BER reaches $10^{-6}$ when SNR increases to $10.55dB$.

$\quad$For 4-PAM, the simulation result of BER reaches $10^{-6}$ when SNR increases to $14.45dB$.

$\quad$The SNR difference between 2-PAM and 4-PAM reach the BER of $10^{-6}$ is $3.9dB$.

$\quad$When the SNR is the same, the 2-PAM BER is always lower than that of 4-PAM, and it becomes more significant when the SNR is larger. The reason is that both 2-PAM and 4-PAM have the same dimension (one-dimension), and 4-PAM carries one more bit than 2-PAM. This ends up 4-PAM has a shorter $d_{min}$, so the BER will be larger.

### c.
***Derive the exact BER and represent it using SNR defined as $\frac{E_b}{N_0}$ when the Gray mapping is applied. Derive a upper bound and a lower bound for the QPSK BER in terms of SNR.***

$\quad \bullet$ The decision metric:

$\quad$Denote the vector form of the symbol by $\vec{s_m}=(S_{mi}, S_{mq})$. Denote the vector form of the received signal by $\vec{r}=(x, y)$. Since all symbols are equiprobable, apply the ML rule: $\hat{s}=\arg {\underset{1\le m\le M}{\min}}{\Vert \vec{r}-\vec{s_m}\Vert}^2=\arg {\underset{1\le m\le M}{\min}}[(x-S_{mi})^2+(y-S_{mq})^2]$.

$\quad$$D_1=\left\{[(x-S_{1i})^2+(y-S_{1q})^2]\lt {\underset{m=2,3,4}{\min}}\{[(x-S_{mi})^2+(y-S_{mq})^2]\}\right\}$<br>
$\qquad$ $=\left\{[(x-S_{1i})^2+(y-S_{1q})^2]\lt [(x-S_{2i})^2+(y-S_{2q})^2]\right\}\bigcup\left\{[(x-S_{1i})^2+(y-S_{1q})^2]\lt [(x-S_{4i})^2+(y-S_{4q})^2]\right\}$<br>
$\qquad$ $=\left\{(x-S_{1i})^2\lt (x-S_{2i})^2\right\}\bigcup\left\{(y-S_{1q})^2\lt (y-S_{4q})^2\right\}$<br>
$\qquad$ $=\left\{|x-S_{1i}|\lt |x-S_{2i}|\right\}\bigcup\left\{|y-S_{1q}|\lt |y-S_{4q}|\right\}$<br>
$\qquad$ $=\{x\gt 0, y\gt 0\}$

$\quad$Similarly,<br>
$\quad$$D_2=\{x\lt 0, y\gt 0\}, D_3=\{x\lt 0, y\lt 0\}, D_4=\{x\gt 0, y\lt 0\}$

$\quad \bullet$ The exact BER:

$\quad$When the Gray mapping is applied, two adjacent symbols only have one bit changes. For instant, if $S_1$ is transmitted, we can know that there is 1 bit error when receiving $S_2$ or $S_4$, and there are 2 bits error when receiving $S_3$. So, the symbol error rate can be specified to bit error rate. Denote the bit SNR $\frac{E_b}{N_0}=\gamma_b$.

$\quad$By symmetric, $P_{b|s_1}=P_{b|s_2}=P_{b|s_3}=P_{b|s_4}$.

$\quad$$P_b=4\times\frac{1}{4}P_{b|s_1}=P_{b|s_1}=\frac{1}{2}\int_{D_2}P(r|s_1)dr+\int_{D_3}P(r|s_1)dr+\frac{1}{2}\int_{D_4}P(r|s_1)dr$<br>
$\qquad$ $=\frac{1}{2}\int_{0}^{\infty}\int_{-\infty}^{0}P(x, y|S_{1i}, S_{1q})dxdy+\int_{-\infty}^{0}\int_{-\infty}^{0}P(x, y|S_{1i}, S_{1q})dxdy+\frac{1}{2}\int_{-\infty}^{0}\int_{0}^{\infty}P(x, y|S_{1i}, S_{1q})dxdy$

$\quad$Since $(x, S_{1i})$ is linear independent with $(y, S_{1q})$ and due to symmetry, $\int_{0}^{\infty}\int_{-\infty}^{0}P(x, y|S_{1i}, S_{1q})dxdy=\int_{-\infty}^{0}\int_{0}^{\infty}P(x, y|S_{1i}, S_{1q})dxdy$, so

$\quad$$P_b=\int_{-\infty}^{0}\int_{0}^{\infty}P(x, y|S_{1i}, S_{1q})dxdy+\int_{-\infty}^{0}\int_{-\infty}^{0}P(x, y|S_{1i}, S_{1q})dxdy$<br>
$\qquad$ $=\int_{-\infty}^{0}\int_{-\infty}^{\infty}P(x, y|S_{1i}, S_{1q})dxdy$<br>
$\qquad$ $=\int_{-\infty}^{0}\int_{-\infty}^{\infty}P(n_1=x-S_{1i}, n_2=y-S_{1q})dxdy$<br>
$\qquad$ $=\int_{-\infty}^{0}\int_{-\infty}^{\infty}\frac{1}{\pi N_0}e^{-\frac{(x-\sqrt{E_b})^2+(y-\sqrt{E_b})^2}{N_0}}dxdy$<br>
$\qquad$ $=\int_{-\infty}^{\infty}\frac{1}{\sqrt{\pi N_0}}e^{-\frac{(x-\sqrt{E_b})^2}{N_0}}dx \int_{-\infty}^{0}\frac{1}{\sqrt{\pi N_0}}e^{-\frac{(y-\sqrt{E_b})^2}{N_0}}dy$<br>
$\qquad$ $=\int_{-\infty}^{\infty}\frac{1}{\sqrt{2\pi}}e^{-\frac{p^2}{2}}dp \int_{-\infty}^{-\sqrt{\frac{2E_b}{N_0}}}\frac{1}{\sqrt{2\pi}}e^{-\frac{q^2}{2}}dq$<br>
$\qquad$ $=1\times [1-Q(-\sqrt{\frac{2E_b}{N_0}})]$<br>
$\qquad$ $=Q(\sqrt{\frac{2E_b}{N_0}})=Q(\sqrt{2\gamma_b})$

$\quad \bullet$ The upper bound and the lower bound:

$\quad$Since the Gray mapping is applied, the number of bit errors can be determined by distinguishing the integration region. However, sometimes it can be wrong. For example, the transmitter send the signal $S_1$, after the AWGN channel it becomes more likely $S_2$, but the estimated signal at the receiver end might be in favour of $S_3$. This will result in the case that the receiver calculates one more bit error, and vice versa.

$\quad$Thus, for the upper bound, regard all regions to have a 2-bit error.<br>
$\quad$And, for the lower bound, regard all regions to have an 1-bit error.<br>
$\quad$It is obvious that $P_{b,upper}=\log_{2}{M}\times P_{b,lower}=2P_{b,lower}$.

$\quad$$P_{b,upper}=4\times\frac{1}{4}P_{b,upper|s_1}=P_{b,upper|s_1}=\int_{D_1^C}P(r|s_1)dr=\int_{D_2}P(r|s_1)dr+\int_{D_3,D_4}P(r|s_1)dr$<br>
$\qquad\quad$ $\space=\int_{0}^{\infty}\int_{-\infty}^{0}P(x, y|S_{1i}, S_{1q})dxdy+\int_{-\infty}^{0}\int_{-\infty}^{\infty}P(x, y|S_{1i}, S_{1q})dxdy$<br>
$\qquad\quad$ $\space=\int_{0}^{\infty}\int_{-\infty}^{0}P(n_1=x-S_{1i}, n_2=y-S_{1q})dxdy+P_b$<br>
$\qquad\quad$ $\space=\int_{0}^{\infty}\int_{-\infty}^{0}\frac{1}{\pi N_0}e^{-\frac{(x-\sqrt{E_b})^2+(y-\sqrt{E_b})^2}{N_0}}dxdy+Q(\sqrt{\frac{2E_b}{N_0}})$<br>
$\qquad\quad$ $\space=\int_{-\infty}^{0}\frac{1}{\sqrt{\pi N_0}}e^{-\frac{(x-\sqrt{E_b})^2}{N_0}}dx \int_{0}^{\infty}\frac{1}{\sqrt{\pi N_0}}e^{-\frac{(y-\sqrt{E_b})^2}{N_0}}dy+Q(\sqrt{\frac{2E_b}{N_0}})$<br>
$\qquad\quad$ $\space=\int_{-\infty}^{-\sqrt{\frac{2E_b}{N_0}}}\frac{1}{\sqrt{2\pi}}e^{-\frac{p^2}{2}}dp \int_{-\sqrt{\frac{2E_b}{N_0}}}^{\infty}\frac{1}{\sqrt{2\pi}}e^{-\frac{q^2}{2}}dq+Q(\sqrt{\frac{2E_b}{N_0}})$<br>
$\qquad\quad$ $\space=\left\{[1-Q(-\sqrt{\frac{2E_b}{N_0}})]\times Q(-\sqrt{\frac{2E_b}{N_0}})\right\}+Q(\sqrt{\frac{2E_b}{N_0}})$<br>
$\qquad\quad$ $\space=\left\{Q(\sqrt{\frac{2E_b}{N_0}})\times [1-Q(\sqrt{\frac{2E_b}{N_0}})]\right\}+Q(\sqrt{\frac{2E_b}{N_0}})$<br>
$\qquad\quad$ $\space=2Q(\sqrt{\frac{2E_b}{N_0}})-Q^2(\sqrt{\frac{2E_b}{N_0}})$<br>
$\qquad\quad$ $\space=2Q(\sqrt{2\gamma_b})-Q^2(\sqrt{2\gamma_b})$

$\quad$$P_{b,lower}=\frac{1}{2}P_{b,upper}=Q(\sqrt{\frac{2E_b}{N_0}})-\frac{1}{2}Q^2(\sqrt{\frac{2E_b}{N_0}})=Q(\sqrt{2\gamma_b})-\frac{1}{2}Q^2(\sqrt{2\gamma_b})$

### d.
***Simulate QPSK modulation and detection in baseband with AWGN to obtain experimental BER using Jupyter notebook or Matlab. Please simulate until the BER reaches $10^{âˆ’6}$ by varying the SNR. Plot the curve of BER vs. SNR. In this plot, the horizontal axis is SNR in dB-scale, and the vertical axis is BER in log-scale.***

```python
# Perform QPSK modulation 00->(1,1) 01->(-1,1) 11->(-1,-1) 10->(1,-1)
def qpsk_gray_modulation(bits, Eb):
    symbols = (1 - 2*bits[1::2]) + 1j * (1 - 2*bits[::2])
    symbols = symbols * np.sqrt(Eb)
    return symbols

# Simulate AWGN channel N(0,N_0/2)
def qpsk_add_awgn_noise(signal, noise_power):
    noise = np.sqrt(noise_power/2) * (np.random.randn(*signal.shape) + 1j * np.random.randn(*signal.shape))
    received_signal = signal + noise
    return received_signal

# Perform QPSK demodulation D1={x>0,y>0}, D2={x<0,y>0}, D={x<0,y<0}, D4={x>0,y<0}
def qpsk_demodulation(received_symbols):
    demodulated_bits = np.zeros(2 * len(received_symbols), dtype=int)
    demodulated_bits[1::2] = np.real(received_symbols) < 0
    demodulated_bits[::2] = np.imag(received_symbols) < 0
    demodulated_bits[1::2][np.real(received_symbols)==0] = random.choices([0, 1], k=1)
    demodulated_bits[::2][np.imag(received_symbols)==0] = random.choices([0, 1], k=1)
    return demodulated_bits

# Calculate error bit number
def error_num(original_bits, received_bits):
    return np.sum(original_bits != received_bits)


# Simulation parameters
Eb = 1.0  # Fixed bit energy
num_bits = 10000  # Number of bits per simulation
error_up_num = 100  # At least error bits to find
qpsk_snr_range_dB = np.append(np.arange(-30, 10.4, 0.2), np.arange(10.4, 11.05, 0.05))  # SNR range in dB
qpsk_snr_range = 10**(qpsk_snr_range_dB/10)  # SNR range
qpsk_noise_range = Eb / qpsk_snr_range  # Noise range

# Initialize arrays to store BER values
ber_values_qpsk = np.zeros(len(qpsk_snr_range_dB))

# Perform simulation for each noise value
def QPSK():
    for i, noise in enumerate(qpsk_noise_range):
        error_count = 0
        loop_count = 0
        while(error_count <= error_up_num):  # Simulate until error bits is sufficient
            bits = np.random.randint(0, 2, num_bits)  # Generate random bits
            modulated_symbols = qpsk_gray_modulation(bits, Eb)
            received_symbols = qpsk_add_awgn_noise(modulated_symbols, noise)
            demodulated_bits = qpsk_demodulation(received_symbols)
            error_count += error_num(bits, demodulated_bits)
            loop_count += 1
            
        # Calculate BER
        ber_values_qpsk[i] = 1.0*error_count / (num_bits*loop_count)

# Simulation
QPSK()
```
